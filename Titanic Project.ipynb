```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
df = pd.read_csv("train.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
le_age= LabelEncoder()
le_cabin = LabelEncoder()
df["Sex"] = le_Sex.fit_transform(df["Sex"])
df["Age"] = le_age.fit_transform(df["Age"])
df["Cabin"]  =le_cabin.fit_transform(df["Cabin"])
newdf = pd.DataFrame()
newdf["PassengerId"] = df["PassengerId"]
newdf["Pclass"] = df["Pclass"]
newdf["Sex"] = df["Sex"]
newdf["Age"] = df["Age"]
newdf["SibSp"] = df["SibSp"]
newdf["Parch"] = df["Parch"]
newdf["Cabin"] = df["Cabin"]
newdff = pd.DataFrame()
newdff["Survived"] = df["Survived"]
```

```python
clf.fit(newdf,newdff)
```

```python
clf.score(newdf,newdff)
```
